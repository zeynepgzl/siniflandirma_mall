```python
import pandas as pd
```

```python
suc_orani = pd.read_csv("suc_orani.csv" , sep=';')
suc_orani['3 kati'] = 3 * suc_orani['sayi']
suc_orani
```

```python
import matplotlib.pyplot as plt

plt.scatter(suc_orani['yil'], suc_orani['3 kati'])
plt.xlabel('Yıl')
plt.ylabel('3 Katı')
plt.title('Yıla Göre 3 Kat')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=26)
kmeans.fit(suc_orani[['yil', '3 kati']])
suc_orani['küme'] = kmeans.predict(suc_orani[['yil', '3 kati']])
suc_orani
```

27 rows × 5 columns

```python
plt.scatter(suc_orani['yil'], suc_orani['3 kati'], c=suc_orani['küme'])
plt.xlabel('Yıl')
plt.ylabel('3 katı')
plt.title('Yıla Göre 3 katı')
plt.show()
```

```python
suc_orani['yil_oran'] = suc_orani['sayi']/365

plt.scatter(suc_orani['yil_oran'], suc_orani['sayi'])
plt.xlabel('Yıllık Oran')
plt.ylabel('Sayı')
plt.title('Yıla Göre Oran')
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=26)
kmeans2.fit(suc_orani[['yil_oran', 'sayi']])
suc_orani['küme2'] = kmeans2.predict(suc_orani[['yil_oran', 'sayi']])
suc_orani
```

```python
plt.scatter(suc_orani['yil_oran'], suc_orani['sayi'], c=suc_orani['küme2'])
plt.xlabel('Oran')
plt.ylabel('Sayı')
plt.title('Yıla Göre Oran')
plt.show()
```

```python
plt.scatter(suc_orani['yil_oran'], suc_orani['sayi'], c=suc_orani['küme2'])
plt.scatter(kmeans2.cluster_centers_[:,0], kmeans2.cluster_centers_[:,1], c='purple')
plt.xlabel('Oran')
plt.ylabel('Sayı')
plt.title('Yıla Göre Oran')
plt.show()
```

```python
suc_orani['TehlikeDurumu'] = suc_orani['küme2'].replace({1: 'çok tehilkeli', 12: 'tehlikeli', 15: 'orta', 20: 'az tehlikeli', 0: 'tehlikesiz'})
suc_orani
```

```python
suc_orani.to_csv('suc_orani_etiketli.csv', index=False)
```
